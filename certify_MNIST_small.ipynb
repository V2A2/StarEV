```python
from StarV.util.load import load_sigmoidal_networks
from StarV.verifier.certifier import certifyRobustness
from tabulate import tabulate
```

```python
data_type = 'mnist'
net_size = 'small'
func = 'tanh'

depth_reduction = 2
relaxation_factor = 0.0

epsilon = 0.02
verification_method = 'BFS'
reachability_method = 'approx'
lp_solver = 'gurobi'

parallel_processing = None
show = True

net, image, label = load_sigmoidal_networks(data_type=data_type, net_size=net_size, func=func)

if show:
    print('epsilon: {}'.format(epsilon))
    print('depth reduction: {}'.format(depth_reduction))
    print('relaxation_factor: {}'.format(relaxation_factor))
    print('verification_method: {}'.format(verification_method))
    print('reachability_method: {}'.format(reachability_method))
    print('lp_solver: {}\n'.format(lp_solver))
    net.info()

rb2, vt2 =certifyRobustness (
    net = net, 
    input = image, 
    epsilon = epsilon, 
    veriMethod = verification_method, 
    reachMethod = reachability_method, 
    lp_solver = lp_solver, 
    pool = parallel_processing, 
    RF = relaxation_factor, 
    DR = depth_reduction, 
    show=show
)
```

```python
print('rb2: ', (rb2==1).sum())
print('vt2: ', vt2.sum())
```

```python
import copy
RF = 0.5
delta = 0.95
layer = 'logsig'

dim = 10
lb = -np.random.rand(dim)
ub = np.random.rand(dim)

lb[np.random.randint(0,dim, 3)] = -0.96
ub[np.random.randint(0,dim, 3)] = 0.96
print('lb: ', lb)
print('ub: ', ub)

l = copy.deepcopy(lb)
u = copy.deepcopy(ub)

print('l.shape[0]: ', l.shape[0])
n1 = round( (1 - RF) * l.shape[0] )
print('n1: ', n1)

midx = np.argsort((u - l))[::-1]
print('midx: ', midx)
if layer in ['logsig', 'tansig']:
    print(layer)
```

```python
print('l[midx]:\n', l[midx])
print('u[midx]:\n', u[midx])
print('diff:\n', u[midx]-l[midx])

midb = np.argwhere((l[midx] >= -delta) & (u[midx] <= delta))
print('midb: \n', midb)

n2 = n1
print('n1: ', n1)
check = midb.flatten().shape[0]
if n2 > check:
    n2 = check
print('n2: ', n2)

mid = midx[midb[0:n2]]
print('l[midb]:\n',l[mid])
print('u[midb]:\n',u[mid])

print('midx[midb[0:n2]]]: ', midx[midb[0:n2]])
print('mid: ', mid)
```
